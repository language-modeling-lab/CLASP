```python
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import brown
import json
from pathlib import Path
from tqdm import tqdm

with open('global_metadata.json', 'r') as f:
    data = json.load(f)
plt.figure(figsize=(12, 8))

nltk.download('brown')
# create a chart of the distribution of the sentences by category
categories = brown.categories()
len_sentences_by_category = {cat: 0 for cat in categories}
for item in tqdm(data):
    if item['category'] in categories:
        len_sentences_by_category[item['category']] += 1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
plt.barh(list(len_sentences_by_category.keys()), list(len_sentences_by_category.values()), color='#d79b00', alpha=0.6)

# Add labels and legend
plt.xlabel('Number of Sentences', fontsize=16)
plt.ylabel('Category', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)  # Increase the font size of category labels
plt.legend(fontsize=14)
plt.title('Sentence Distribution by Category for Speech Brown Dataset', fontsize=16)

# Show the plot
plt.tight_layout()
plt.savefig('sentence_distribution_speech_brown.png', dpi=300, transparent=True)  # Save as high-quality image
plt.show()
```
